```python
# Dropping the columns not needed
df2 = df1.drop(['Address', 'dateSold', 'daysOnZillow', 'homeStatus', 'homeType', 'lastSoldPrice',
       'latitude', 'longitude', 'lotSize', 'pageViewCount', 'priceHistory/0/date', 'priceHistory/0/event',
       'priceHistory/0/postingIsRental', 'priceHistory/0/price',
       'priceHistory/0/priceChangeRate', 'priceHistory/0/pricePerSquareFoot',
       'rentZestimate', 'resoFacts/atAGlanceFacts/0/factValue',
       'resoFacts/basement', 'resoFacts/heating/0',
       'resoFacts/atAGlanceFacts/2/factLabel',
       'resoFacts/atAGlanceFacts/2/factValue', 'resoFacts/cooling/0',
       'resoFacts/atAGlanceFacts/3/factLabel',
       'resoFacts/atAGlanceFacts/3/factValue', 'taxAssessedValue',
       'taxAssessedYear', 'taxHistory/0/taxIncreaseRate',
       'taxHistory/0/taxPaid', 'taxHistory/0/time', 'taxHistory/0/value',
       'taxHistory/0/valueIncreaseRate', 'url', 'yearBuilt' , 'zestimateLowPercent', 'zpid'],axis='columns')
df2.shape
```

```python
# Displaying the top records
df2.head()
```

```python
# Dropping more columns not needed
df3 = df2.drop(['Unnamed: 0', 'zestimate' , 'zestimateHighPercent'],axis='columns')
df3.head()
```

```python
# Looking for null values
df3.isnull().sum()
```

```python
# Dropping null values
df4 = df3.dropna()
df4.isnull().sum()
```

```python
# Getting the amount of rows and columns
df4.shape
```

```python
# create function to change livingArea to number
def convert_sqft_to_num(x):
    try:
        return float(x)
    except:
        return None
```

```python
df5 = df4.copy()
df5.livingArea = df5.livingArea.apply(convert_sqft_to_num)
df5 = df5[df5.livingArea.notnull()]
df5.head(2)
```

```python
# Creating a new column and creating forumla for price per square foot
df6 = df5.copy()
df6['price_per_sqft'] = df6['price']/df6['livingArea']
df6.head()
```

```python
df6.head(50)
```

```python
df6.location = df6.City.apply(lambda x: x.strip())
location_stats = df6['City'].value_counts(ascending=False)
location_stats
```

```python
location_stats.values.sum()
```

```python
len(location_stats[location_stats>10])
```

```python
df7 = df6.dropna()
df7.isnull().sum()
```

```python
df7.shape
```

```python
#Found this code on stackoverflow to correct an error later in code when running regression algorithum
df7 =df7[~df7.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
X = df7.drop(['price', 'City'],axis='columns')
X.head(3)
```

```python
X.shape
```

```python
y = df7.price
y.head(3)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

# Use K Fold cross validation to measure accuracy of our LinearRegression model

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```
