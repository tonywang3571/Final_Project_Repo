```python
import requests
import json
import pandas as pd

# position stack API key
import sys
sys.path.append("C:/Users/laura/bootcamp/")
from api_keys import position_stack
from api_keys import g_key
```

```python
zillow_df_google=pd.read_csv('Resources/cleaned_zillow.csv')
zillow_df_google
```

```python
zillow_df_google["house_number"]=""
zillow_df_google["street_name"]=""
zillow_df_google["zipcode"]=""
zillow_df_google["neighborhood"]=""
zillow_df_google
```

```python
#create a lat, long string column to pass into API
zillow_df_google["lat_lng"]=zillow_df_google["latitude"].astype(str) + ', ' + zillow_df_google["longitude"].astype(str)
zillow_df_google
```

```python
# create a params dict that will be updated with new address iteration
params = {"key": g_key}

# Loop through the df and run address search for each lat_lng pair
for index, row in zillow_df_google.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    #get lat_lng from df
    lat_lng=row['lat_lng']

    # update lat,lng string to key value
    params['latlng'] = lat_lng

    # make request and convert to json
    response = requests.get(base_url, params=params).json()
    
    #extract results
    results=response["results"]    
    
 
    #extract results
    try:
        zillow_df_google.loc[index, "neighborhood"] = results[0]["address_components"][2]["long_name"]
        zillow_df_google.loc[index, "zipcode"] = results[0]["address_components"][7]["long_name"]
        zillow_df_google.loc[index, "house_number"] = results[0]["address_components"][0]["long_name"]
        zillow_df_google.loc[index, "street_name"] = results[0]["address_components"][1]["long_name"]
    except (IndexError):
        print("Not Found Skip")
    
#     arbnb_df.loc[index, "Neighborhood"] = results["results"][0]["address_components"][2]["long_name"]
#     arbnb_df.loc[index, "Zipcode"] = results["results"][0]["address_components"][7]["long_name"]
#     arbnb_df.loc[index, "house_number"] = results["results"][0]["address_components"][0]["long_name"]
#     arbnb_df.loc[index, "street_name"] = results["results"][0]["address_components"][1]["long_name"]
  

# Visualize to confirm lat lng appear
zillow_df_google
```
